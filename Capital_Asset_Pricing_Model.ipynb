# CALCULATING EXPECTED RETURN USING CAPM

```python
def calculate_expected_return(Rf, beta, E_rm):
    # CAPM formula
    expected_return = Rf + beta * (E_rm - Rf)
    return expected_return

Rf = 0.03  # Risk-free rate (as a decimal)
beta = 1.2  # Beta of the asset
E_rm = 0.1  # Expected return on the market (as a decimal)


expected_return = calculate_expected_return(Rf, beta, E_rm)

print(f"Expected Return: {expected_return:.2%}")
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


class CAPMCalculator:
    def __init__(self, stock_symbols, start_date, end_date, risk_free_rate, expected_market_return):
        self.stock_symbols = stock_symbols
        self.start_date = start_date
        self.end_date = end_date
        self.risk_free_rate = risk_free_rate
        self.expected_market_return = expected_market_return
        self.beta_results = pd.DataFrame(columns=['Stock Symbol', 'Beta'])
        self.capm_results = pd.DataFrame(columns=['Stock Symbol', 'Beta', 'Expected Return (CAPM)'])

    def download_historical_data(self, symbol):
        data = yf.download(symbol, start=self.start_date, end=self.end_date)
        data['Daily_Return'] = data['Adj Close'].pct_change()
        data = data.dropna()
        return data

    def calculate_beta(self, stock_symbol, market_data):
        stock_data = self.download_historical_data(stock_symbol)
        combined_data = pd.concat([stock_data['Daily_Return'], market_data['Daily_Return']], axis=1)
        combined_data.columns = [stock_symbol, 'Market']
        covariance_matrix = np.cov(combined_data[stock_symbol], combined_data['Market'])
        beta = covariance_matrix[0, 1] / covariance_matrix[1, 1]
        return beta

    def calculate_expected_returns(self):
        market_data = yf.download('^GSPC', start=self.start_date, end=self.end_date)
        market_data['Daily_Return'] = market_data['Adj Close'].pct_change()
        market_data = market_data.dropna()

        for symbol in self.stock_symbols:
            beta = self.calculate_beta(symbol, market_data)
            expected_return = self.risk_free_rate + beta * (self.expected_market_return - self.risk_free_rate)

            self.beta_results = self.beta_results.append({'Stock Symbol': symbol, 'Beta': beta}, ignore_index=True)
            self.capm_results = self.capm_results.append({'Stock Symbol': symbol, 'Beta': beta, 'Expected Return (CAPM)': expected_return}, ignore_index=True)

    def display_beta_results(self):
        print("Beta values for the selected stocks:")
        print(self.beta_results)

    def display_capm_results(self):
        print("Expected Returns (CAPM) for the selected stocks:")
        print(self.capm_results)

    def plot_stock_and_market_returns(self, stock_symbol):
        stock_data = self.download_historical_data(stock_symbol)
        market_data = yf.download('^GSPC', start=self.start_date, end=self.end_date)
        market_data['Daily_Return'] = market_data['Adj Close'].pct_change()
        market_data = market_data.dropna()

        plt.figure(figsize=(12, 6))
        plt.plot(stock_data.index, stock_data['Daily_Return'], label=f'{stock_symbol} Returns')
        plt.plot(market_data.index, market_data['Daily_Return'], label='Market Returns')
        plt.title(f'Daily Returns for {stock_symbol} and Market')
        plt.xlabel('Date')
        plt.ylabel('Daily Return')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_capm_results(self):
        plt.figure(figsize=(12, 6))
        plt.bar(self.capm_results['Stock Symbol'], self.capm_results['Expected Return (CAPM)'])
        plt.title('Expected Returns (CAPM) for the selected stocks')
        plt.xlabel('Stock Symbol')
        plt.ylabel('Expected Return (CAPM)')
        plt.xticks(rotation=45)
        plt.grid(axis='y')
        plt.show()

if __name__ == "__main__":
    stock_symbols = ['AAPL', 'GOOGL', 'AMZN', 'TSLA', 'MSFT', 'BRK-B', 'NVDA', 'JPM', 'JNJ', 'NFLX']
    start_date = "2022-01-01"
    end_date = "2023-01-01"
    risk_free_rate = 0.025  # 10-year U.S. Treasury yield
    expected_market_return = 0.1  # Expected return for the market

    capm_calculator = CAPMCalculator(stock_symbols, start_date, end_date, risk_free_rate, expected_market_return)
    capm_calculator.calculate_expected_returns()
    capm_calculator.display_beta_results()
    capm_calculator.display_capm_results()

    # Plot individual stock and market returns for a selected stock (e.g., AAPL)
    capm_calculator.plot_stock_and_market_returns('AAPL')
    capm_calculator.plot_stock_and_market_returns('NVDA')
    capm_calculator.plot_stock_and_market_returns('TSLA')
    capm_calculator.plot_stock_and_market_returns('AMZN')
    capm_calculator.plot_stock_and_market_returns('NFLX')
    # Plot CAPM results for expected returns
    capm_calculator.plot_capm_results()
```
